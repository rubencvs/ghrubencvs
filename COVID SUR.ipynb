```python
# Define a map centered around Santiago
santiago_map = folium.Map(location=[-33.58472935,-70.67480915038735], zoom_start=14)

#Mark the center
for index, row in pino_data.iterrows():
    linea = row['point']
    labelpp = row['ID']
    linea = linea.replace("(", "") 
    linea = linea.replace(")", "")
    loglat = linea.split(",")   
    micolor='red'
    deceased = row['PAPMIDeceased']
    if deceased=='Y':
        micolor = 'black'
    folium.CircleMarker(
                [loglat[0],loglat[1]],
                radius=3,
                color=micolor,
                fill=True,
                popup=labelpp,
                fill_color='darkred',
                fill_opacity=0.6
            ).add_to(santiago_map)

    
santiago_map
```
